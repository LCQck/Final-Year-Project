```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Load the CSV file
data = pd.read_csv('/content/Impact_of_Remote_Work_on_Mental_Health.csv')

# Display first few rows to check data
print(data.head())

# Visualization of 'Hours Worked Per Week' vs 'Stress Level' (assuming this is a mental health measure)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_Worked_Per_Week', y='Stress_Level', data=data)
plt.title('Hours Worked Per Week vs Stress Level')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Stress Level')
plt.show()

# Visualization of 'Work Location' vs 'Stress Level'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Work_Location', y='Stress_Level', data=data)
plt.title('Work Location vs Stress Level')
plt.xlabel('Work Location')
plt.ylabel('Stress Level')
plt.show()

# Multiple Regression Analysis
# Define the formula for multiple regression
formula = 'Stress_Level ~ Hours_Worked_Per_Week + C(Work_Location)'

# Fit the regression model
model = smf.ols(formula, data=data).fit()

# Output model summary
print(model.summary())

# Plot regression residuals to verify model assumptions
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(model, 'Hours_Worked_Per_Week', fig=fig)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Load the CSV file
data = pd.read_csv('/content/Impact_of_Remote_Work_on_Mental_Health.csv')

# Convert 'Stress_Level' to numeric values
stress_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
data['Stress_Level_Numeric'] = data['Stress_Level'].map(stress_mapping)

# Check if mapping was successful
print(data[['Stress_Level', 'Stress_Level_Numeric']].head())

# Visualization of 'Hours Worked Per Week' vs 'Stress Level'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_Worked_Per_Week', y='Stress_Level_Numeric', data=data)
plt.title('Hours Worked Per Week vs Stress Level')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Stress Level (Numeric)')
plt.show()

# Visualization of 'Work Location' vs 'Stress Level'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Work_Location', y='Stress_Level_Numeric', data=data)
plt.title('Work Location vs Stress Level')
plt.xlabel('Work Location')
plt.ylabel('Stress Level (Numeric)')
plt.show()

# Multiple Regression Analysis
# Define the formula for multiple regression
formula = 'Stress_Level_Numeric ~ Hours_Worked_Per_Week + C(Work_Location)'

# Fit the regression model
model = smf.ols(formula, data=data).fit()

# Output model summary
print(model.summary())

# Plot regression residuals to verify model assumptions
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(model, 'Hours_Worked_Per_Week', fig=fig)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import AnovaRM

# Load the CSV file
data = pd.read_csv('/content/Impact_of_Remote_Work_on_Mental_Health.csv')

# Convert 'Productivity_Change' to a numeric scale
prod_change_mapping = {'Decrease': -1, 'No Change': 0, 'Increase': 1}
data['Productivity_Change_Numeric'] = data['Productivity_Change'].map(prod_change_mapping)

# Perform Two-Way ANOVA
formula = 'Productivity_Change_Numeric ~ C(Access_to_Mental_Health_Resources) * C(Mental_Health_Condition)'
model = smf.ols(formula, data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualization 1: Boxplot for Productivity Change by Access to Mental Health Resources and Mental Health Condition
plt.figure(figsize=(12, 6))
sns.boxplot(x='Access_to_Mental_Health_Resources', y='Productivity_Change_Numeric', hue='Mental_Health_Condition', data=data)
plt.title('Impact of Access to Mental Health Resources and Mental Health Condition on Productivity Change')
plt.xlabel('Access to Mental Health Resources')
plt.ylabel('Productivity Change')
plt.legend(title='Mental Health Condition')
plt.show()

# Visualization 2: Interaction Plot
from statsmodels.graphics.factorplots import interaction_plot

plt.figure(figsize=(10, 6))
interaction_plot(data['Access_to_Mental_Health_Resources'], data['Mental_Health_Condition'], data['Productivity_Change_Numeric'],
                 colors=['red', 'blue', 'green'], markers=['D', '^', 'o'], ms=10)
plt.title('Interaction between Access to Mental Health Resources and Mental Health Condition on Productivity Change')
plt.xlabel('Access to Mental Health Resources')
plt.ylabel('Mean Productivity Change')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('/content/Impact_of_Remote_Work_on_Mental_Health.csv')

# Map 'Productivity_Change' to numeric values for increase (+1) and decrease (-1)
prod_change_mapping = {'Decrease': -1, 'No Change': 0, 'Increase': 1}
data['Productivity_Change_Numeric'] = data['Productivity_Change'].map(prod_change_mapping)

# Combined effect: Calculate mean productivity change for each combination of Mental Health Condition and Access to Resources
combined_pivot_table = data.pivot_table(values='Productivity_Change_Numeric',
                                        index='Mental_Health_Condition',
                                        columns='Access_to_Mental_Health_Resources',
                                        aggfunc='mean')

# Separate effects: Average Productivity Change by Mental Health Condition
mental_health_condition_pivot = data.pivot_table(values='Productivity_Change_Numeric',
                                                 index='Mental_Health_Condition',
                                                 aggfunc='mean')

# Average Productivity Change by Access to Mental Health Resources
access_to_resources_pivot = data.pivot_table(values='Productivity_Change_Numeric',
                                             index='Access_to_Mental_Health_Resources',
                                             aggfunc='mean')

# Plot combined effect heatmap with red and blue colors
plt.figure(figsize=(8, 6))
sns.heatmap(combined_pivot_table, annot=True, cmap="coolwarm", center=0, cbar_kws={'label': 'Average Productivity Change'}, vmin=-1, vmax=1)
plt.title('Combined Effect of Access to Resources and Mental Health Condition on Productivity Change')
plt.xlabel('Access to Mental Health Resources')
plt.ylabel('Mental Health Condition')
plt.show()

# Plot Mental Health Condition effect heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(mental_health_condition_pivot.T, annot=True, cmap="coolwarm", center=0, cbar_kws={'label': 'Average Productivity Change'}, vmin=-1, vmax=1)
plt.title('Effect of Mental Health Condition on Productivity Change')
plt.xlabel('Mental Health Condition')
plt.ylabel('')
plt.show()

# Plot Access to Resources effect heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(access_to_resources_pivot.T, annot=True, cmap="coolwarm", center=0, cbar_kws={'label': 'Average Productivity Change'}, vmin=-1, vmax=1)
plt.title('Effect of Access to Mental Health Resources on Productivity Change')
plt.xlabel('Access to Mental Health Resources')
plt.ylabel('')
plt.show()
```
